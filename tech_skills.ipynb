```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("dataset.csv")
df
```

```python
titles_raw = df["Title"]
titles_raw
titles = []
for x in titles_raw:
    if x[0] == '[':
        titles.append(x[2:len(x)-2])
    elif x[0] == '*':
        titles.append(x[1:len(x)-1])
    else:
        titles.append(x)
titles
```

```python
# categories = ["manager", "specialist", "engineer", 
import nltk

text = ""
for item in titles: 
#     if '/' in item:
#         item = item.replace('/', '')
#     if ',' in item:
#         item = item.replace(',', '')
    text += f'{item} '
text
categories = nltk.tokenize.word_tokenize(text)
categories

# categoriesDist = nltk.FreqDist(x.lower() for x in categories)
# categoriesDist

stopwords = nltk.corpus.stopwords.words('english')
categoriesExceptStopDist = nltk.FreqDist(x.lower() for x in categories if x.lower() not in stopwords)    
categoriesExceptStopDist

mostCommon = categoriesExceptStopDist.most_common(40)
mostCommon
```

```python
# job_title_categories = []
# for i in mostCommon:
#     job_title_categories.append(i[0])
# print(job_title_categories)

# Manually filtered/shortlisted categories
job_title_categories = ['penetration', 'tester', 'senior', 'security', 'remote', 'clearance', 'secret', 'cyber', 'testing', 'engineer', 'team', 'application', 'lead', 'consultant', 'red', 'vulnerability', 'cybersecurity', 'specialist', 'web', 'junior', 'analyst', 'pen', 'manager', 'pentester', 'engineer/', 'testers', 'network', 'technology', 'consulting']
job_title_categories = job_title_categories[::-1]
print(job_title_categories)
```

```python
# Distribute jobs titles in categories
# From the least popular categories to most popular
new_column = []
for index, row in df.iterrows():
    job_title = row['Title'].lower()
    if 'technology' in job_title:
        new_column.append('technology')
    elif 'network' in job_title:
        new_column.append('network')
    elif 'manager' in job_title:
        new_column.append('manager')
    elif 'analyst' in job_title:
        new_column.append('analyst')
    elif 'junior' in job_title:
        new_column.append('junior')
    elif 'web' in job_title:
        new_column.append('web')
    elif 'specialist' in job_title:
        new_column.append('specialist')
    elif 'vulnerability' in job_title:
        new_column.append('vulnerability')
    elif 'red' in job_title:
        new_column.append('red')
    elif ('consulting' in job_title) or ('consultant' in job_title):
        new_column.append('consultant')
    elif 'lead' in job_title:
        new_column.append('lead')
#     elif 'application' in job_title:
#         new_column.append('application')
    elif 'team' in job_title:
        new_column.append('team')
    elif ('engineer' in job_title) or ('engineer/' in job_title):
        new_column.append('engineer')
    elif 'secret' in job_title:
        new_column.append('secret')
    elif 'clearance' in job_title:
        new_column.append('clearance')
    elif 'remote' in job_title:
        new_column.append('remote')
    elif ('senior' in job_title) or ('sr.' in job_title):
        new_column.append('senior')
    elif ('cybersecurity' in job_title) or ('cyber' in job_title) or ('security' in job_title):
        new_column.append('security')
    elif ('tester' in job_title) or ('testers' in job_title) or ('pentester' in job_title) or ('pen' in job_title) or ('penetration' in job_title)  or ('testing' in job_title):
        new_column.append('pentester')
    else: 
        new_column.append('not-specific')
new_column
```

```python
df['Title-Category'] = new_column
df
```

```python
categories = new_column
print(list(pd.unique(categories)))
```

```python
tech_skills = ['automated tools', 'penetration Test', 'manual code review', 'threat modelling', 'application business logic', 'application architecture', 'sast analysis', 'false positive analysis', 'testing scenarios', 'document', 'manipulation of network', 'shell scripting', 'exploits', 'revere engineering', 'source code review', 'network protocols', 'offensive security', 'vulnerability identification', 'vulnerability detection', 'vulnerability prediction', 'system design review', 'code analysis', 'vulnerability exploit', 'architecture reviews', 'cloud penetration', 'cryptography', 'security review audit', 'review security policies', 'public accounting', 'firewall', 'router', 'switch', 'cryptographic', 'vulnerability modelling', 'mitigate cyber risk', 'Nessus', 'Federal security regulations', 'nist', 'vendor report review', 'process review', 'simulation', 'simulate', 'social engineering', 'security audit', 'application development', 'osi layers', 'cloud security', 'vulnerability scanners', 'exploitation frameworks', 'manual testing', 'static code analysis', 'dynamic code analysis', 'owasp', 'network sniffing', 'active directory analysis', 'firewall rule analysis', 'IPS/IDS', 'cyber kill']

title_categories_list = ['pentester', 'specialist', 'engineer', 'lead', 'consultant', 'junior', 'security', 'web', 'senior', 'analyst', 'manager', 'red', 'vulnerability', 'technology', 'clearance', 'remote', 'team', 'secret', 'network']

data = {'Tech_Skills':tech_skills}
new_df = pd.DataFrame(data=data)

for category in title_categories_list: 
    temp_df = df.loc[df["Title-Category"] == category]
    tech_skills_dict = {skill.lower():0 for skill in tech_skills}
    for index, row in temp_df.iterrows():
        job_description = row["Description"]
        for key, value in tech_skills_dict.items():
            if key in job_description.lower():
                tech_skills_dict[key] += 1
    count_column=[value for key,value in tech_skills_dict.items()]
    new_df[category] = count_column

new_df
```

```python
new_df.plot.barh(x='Tech_Skills', y='pentester', rot=0)
```

```python
# drop rows with zeros only
new_df = new_df.drop(index=1)
new_df = new_df.set_index('Tech_Skills')
new_df = new_df.loc[~new_df.apply(lambda row: (row==0).all(), axis=1)]
```

```python
import seaborn as sns
from matplotlib import pyplot as plt
sns.set(style="white")
import itertools
```

```python
fig, axes = plt.subplots(19, 1, sharex=True, figsize=(15,18))
fig.suptitle('Tech Skills VS Job Categories')

# set palette 
palette = itertools.cycle(sns.color_palette())

idx = 0
for item in title_categories_list[::-1]:
    c = next(palette)
    sns.barplot(ax=axes[idx], x=new_df.index, y=new_df[item], color=c)
    axes[idx].set_title(item, fontsize=15, loc='left',pad=-14)
    axes[idx].set(xlabel=None, ylabel='frequency') 
    if idx == 18:
        axes[idx].set_xticklabels(
            axes[idx].get_xticklabels(), 
            rotation=45, 
            horizontalalignment='right',
            fontweight='light',
            fontsize=14
        )
    idx+=1
```
