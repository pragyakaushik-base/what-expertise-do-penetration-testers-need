```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("dataset_job_categories.csv")
df
```

```python
soft_skills = ['analytical', 'collaboration', 'team', 'communication', 'creativity', 'diversity', 'empathy', 'independence', 'interpersonal', 'leadership', 'listening', 'mentoring', 'negotiation', 'problem solving', 'autonomy', 'presentation', 'management', 'responsibility', 'critical think', 'ethic', 'flexibility', 'organizational', 'planning', 'time management', 'commitment', 'eager to learn', 'customer orient', 'motivation']
```

```python
title_categories_list = ['pentester', 'specialist', 'engineer', 'lead', 'consultant', 'junior', 'security', 'web', 'senior', 'analyst', 'manager', 'red', 'vulnerability', 'technology', 'clearance', 'remote', 'team', 'secret', 'network']
```

```python
data = {'Soft_Skills': soft_skills}
new_df = pd.DataFrame(data=data)

for category in title_categories_list: 
    temp_df = df.loc[df["Title-Category"] == category]
    soft_skills_dict = {skill.lower():0 for skill in soft_skills}
    for index, row in temp_df.iterrows():
        job_description = row["Description"]
        for key, value in soft_skills_dict.items():
            if key in job_description.lower():
                soft_skills_dict[key] += 1
    count_column=[value for key,value in soft_skills_dict.items()]
    new_df[category] = count_column

new_df
```

```python
# drop rows with zeros only
new_df = new_df.set_index('Soft_Skills')
new_df = new_df.loc[~new_df.apply(lambda row: (row==0).all(), axis=1)]
new_df
```

```python
import seaborn as sns
from matplotlib import pyplot as plt
sns.set(style="white")
import itertools
```

```python
fig, axes = plt.subplots(19, 1, sharex=True, figsize=(15,18))
fig.suptitle('Soft Skills VS Job Categories')

# set palette 
palette = itertools.cycle(sns.color_palette())

idx = 0
for item in title_categories_list[::-1]:
    c = next(palette)
    sns.barplot(ax=axes[idx], x=new_df.index, y=new_df[item], color=c)
    axes[idx].set_title(item, fontsize=15, loc='left',pad=-14)
    axes[idx].set(xlabel=None, ylabel='frequency') 
    if idx == 18:
        axes[idx].set_xticklabels(
            axes[idx].get_xticklabels(), 
            rotation=45, 
            horizontalalignment='right',
            fontweight='light',
            fontsize=14
        )
    idx+=1
```
